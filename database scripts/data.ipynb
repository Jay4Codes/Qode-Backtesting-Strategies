```python
import pandas as pd
import os
import glob
```

```python
path = "../data/"
all_files = glob.glob(os.path.join(path, "nifty_*.csv"))

for file in all_files:
    print(file)
    df = pd.read_csv(file)
    symbols = df["symbol"].unique()

    symbols = [
        symbol for symbol in symbols if symbol != "NIFTY" and symbol != "NIFTY-I"
    ]

    expiry_dates = [symbol[5:12] for symbol in symbols]

    for expiry_date in expiry_dates:
        for symbol in symbols:
            if symbol[5:12] == expiry_date:
                strike_price = symbol[12:17]
                os.makedirs(
                    f"../database/options/{expiry_date}/{strike_price}", exist_ok=True
                )

    for symbol in symbols:
        expiry_date = symbol[5:12]
        strike_price = symbol[12:17]
        option_type = symbol[17:19]
        df_filtered = df[
            df["symbol"].str.contains(expiry_date + strike_price + option_type)
        ]

        df_filtered = df_filtered.drop(df_filtered.columns[0], axis=1)

        df_filtered["timestamp"] = pd.to_datetime(
            df_filtered["date"] + " " + df_filtered["time"]
        )

        df_filtered = df_filtered.drop(["date", "time"], axis=1)

        df_filtered["timestamp"] = pd.to_datetime(df_filtered["timestamp"])

        df_filtered = df_filtered.sort_values("timestamp")

        if not os.path.exists(
            f"../database/options/{expiry_date}/{strike_price}/{symbol}.parquet"
        ):
            df_filtered.to_parquet(
                f"../database/options/{expiry_date}/{strike_price}/{symbol}.parquet",
                index=False,
            )
        else:
            df_existing = pd.read_parquet(
                f"../database/options/{expiry_date}/{strike_price}/{symbol}.parquet"
            )

            df_merged = pd.concat([df_existing, df_filtered]).drop_duplicates()
            df_merged.to_parquet(
                f"../database/options/{expiry_date}/{strike_price}/{symbol}.parquet",
                index=False,
            )
```
